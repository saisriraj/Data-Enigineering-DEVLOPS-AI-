### EDA (Exploratory Data Analysis)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram: Subscription Length
plt.figure(figsize=(8, 5))
sns.histplot(df['subscription_length'], bins=20, kde=True, color='skyblue')
plt.title('Subscription Length Distribution')
plt.xlabel('Subscription Length')
plt.ylabel('Frequency')
plt.show()

# Count Plot: Fuel Type
plt.figure(figsize=(6, 4))
sns.countplot(x='fuel_type', data=df, palette='Set2')
plt.title('Fuel Type Frequency')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()

# Box Plot: Customer Age
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['customer_age'], color='lightgreen')
plt.title('Customer Age Distribution')
plt.xlabel('Customer Age')
plt.show()
```

```python
# Scatter Plot: Subscription Length vs Customer Age
plt.figure(figsize=(8, 5))
sns.scatterplot(x='subscription_length', y='customer_age', data=df, hue='claim_status', palette='coolwarm')
plt.title('Subscription Length vs Customer Age by Claim Status')
plt.xlabel('Subscription Length')
plt.ylabel('Customer Age')
plt.show()

# Box Plot: Claim Status across Fuel Type
plt.figure(figsize=(6, 4))
sns.boxplot(x='fuel_type', y='customer_age', hue='claim_status', data=df, palette='Set1')
plt.title('Customer Age by Fuel Type and Claim Status')
plt.xlabel('Fuel Type')
plt.ylabel('Customer Age')
plt.show()

# Heatmap: Correlation Matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Feature Correlation Heatmap')
plt.show()
```

Trend & Time-Style Visualizations
Even if you don’t have timestamped data, continuous features can mimic time-like behavior:

🔹 Line Plot: Average Customer Age by Subscription Lengt

```python
# Grouping data
trend_df = df.groupby('subscription_length')['customer_age'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x='subscription_length', y='customer_age', data=trend_df, marker='o', color='navy')
plt.title('Average Customer Age across Subscription Length')
plt.xlabel('Subscription Length')
plt.ylabel('Average Customer Age')
plt.grid(True)
plt.show()
```

```python
# Creating claim rate
claim_rate = df.groupby('vehicle_age')['claim_status'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x='vehicle_age', y='claim_status', data=claim_rate, marker='s', color='darkgreen')
plt.title('Claim Rate across Vehicle Age')
plt.xlabel('Vehicle Age')
plt.ylabel('Claim Rate')
plt.grid(True)
plt.show()
```

```python
# Grouping data
trend_df = df.groupby('subscription_length')['customer_age'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x='subscription_length', y='customer_age', data=trend_df, marker='o', color='navy')
plt.title('Average Customer Age across Subscription Length')
plt.xlabel('Subscription Length')
plt.ylabel('Average Customer Age')
plt.grid(True)
plt.show()
```
